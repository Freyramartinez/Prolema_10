# WHAT OTHER LANGUAGES ARE SPOKEN IN MEXICO?

```python
marcas_clases=["1M", "2M", "3M", "4M", "5M"]
frecuencias=[38626, 695268, 494413, 1575941, 4920952]
marcas_texto=["Romance", "Others", "Italian", "French", "German"]
plt.figure(figsize = (12, 6))
plt.barh(marcas_clases, frecuencias, 
        height = 0.75, edgecolor = "k",
       color = ["#A7F100", "#0724F1", "#CD61CD", "#CD617D", "#61CDBB"])
plt.yticks(marcas_clases, marcas_texto, fontsize = 20)
plt.xlabel("MILLIONS", fontsize = 20)
plt.ylabel("LANGUAGES", fontsize = 20)
plt.title("Switzerland", fontsize = 25)
plt.grid()
plt.show()
```

```python
datos = [63.9, 20.5, 6.4, 0.5, 9 ]
marcas_texto = ["German", "French", "Italian", "Romance", "Others"]
separaciones = [0.1, 0.0, 0, 0, 0]

plt.pie(datos,
       explode = separaciones, 
       counterclock = False,
       startangle = 90, 
       autopct = "%0.1f%%",
       pctdistance = 0.7,
       colors = ["#A7F107", "#0724F1", "#CD61CD", "#CD617D", "#61CDBB"],
       labels = marcas_texto)

plt.title("Languages spoken in Switzerland", fontsize = 25) # Titulo del grafico.
plt.show()
```
