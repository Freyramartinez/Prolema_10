# MOST IMPORTANT LANGUAGES IN DIFERENT COUNTRIES.

### we can represent the diferent languages spoken in any country, in this case, we gonna use the data from Switzerland and México.

#### Of the 7 millions of inhabitants that switzerland have, 21% of them are foreing. As an example of its diversity, four official languages are spoken in Switzerland
#### In Switzerland, the most widely spoken language is German with 63.9% of speakers, followed by French with 20.5% and Italian with 6.4%. Romance languages represent only 0.5% of use. Other languages are used by 9% of the population.

|Language  | Speakers |
|--------|------------|
|Alemán  | 4 920 952  |
|Frances | 1 575 941  |
|Italiano|   494 413  |
|Otros   |   695 268  |
|Romance |    38 625  |

```python
marcas_clases=["1M", "2M", "3M", "4M", "5M"]
frecuencias=[38626, 695268, 494413, 1575941, 4920952]
marcas_texto=["Romance", "Others", "Italian", "French", "German"]
plt.figure(figsize = (12, 6))
plt.barh(marcas_clases, frecuencias, 
        height = 0.75, edgecolor = "k",
       color = ["#A7F100", "#0724F1", "#CD61CD", "#CD617D", "#61CDBB"])
plt.yticks(marcas_clases, marcas_texto, fontsize = 20)
plt.xlabel("MILLIONS", fontsize = 20)
plt.ylabel("LANGUAGES", fontsize = 20)
plt.title("Switzerland", fontsize = 25)
plt.grid()
plt.show()
```

#### Bar graph representing the moat spoken languages in Switzerland by the official percentage

```python
datos = [63.9, 20.5, 6.4, 0.5, 9 ]
marcas_texto = ["German", "French", "Italian", "Romance", "Others"]
separaciones = [0.1, 0.0, 0, 0, 0]

plt.pie(datos,
       explode = separaciones, 
       counterclock = False,
       startangle = 90, 
       autopct = "%0.1f%%",
       pctdistance = 0.7,
       colors = ["#A7F107", "#0724F1", "#CD61CD", "#CD617D", "#61CDBB"],
       labels = marcas_texto)

plt.title("Languages spoken in Switzerland", fontsize = 25) # Titulo del grafico.
plt.show()
```

#### Pie chart representig the percentage of the most important languages spoken in Switzerland

# WHAT OTHER LANGUAGES ARE SPOKEN IN MEXICO?

### In Mexico, Spanish is spoken by 98% of the population, making it the most widely used language. It is followed by Nahuatl, with 1.7% of speakers, and Maya, with 0.7% of Mexican speakers.

|LANGUAGES  | SPEAKERS  |
|--------|------------|
|Español |      98%   |
|Nahuatl |     1.7%   |
|Maya   |     0.7%   |
|Mixteco |     0.6%   |
|Zapoteco|     0.4%   |

```python
marcas_clases=[20, 40, 60, 80, 100]
frecuencias=[0.4, 0.6, 0.7, 1.7, 98]
marcas_texto= ["Zapoteco","Mixtec","Maya","Nahuatl","Spanish"]

plt.figure(figsize = (12, 6))
plt.barh(marcas_clases, frecuencias, 
        height = 5, edgecolor = "k",
       color = ["#A7F100", "#0724F1", "#CD61CD", "#CD617D", "#61CDBB"])
plt.yticks(marcas_clases, marcas_texto, fontsize = 20)
plt.xlabel("PERCENTAGE", fontsize = 20)                                                                                                                                                                                                  
plt.ylabel("LANGUAGES", fontsize = 20)
plt.title("México", fontsize = 25)
plt.grid()
plt.show()
```

#### Bar graph representing the moat spoken languages in Switzerland by the official percentage

```python
datos = [98, 1.7, 0.7, 0.6, 0.4]
marcas_texto= ["Spanish","Nahuatl","Maya","Mixtec","Zapoteco"]
separaciones = [0.1, 0.5, 0.5, 0.4, 0.3]

plt.pie(datos,
       explode = separaciones, 
       counterclock = False,
       startangle = 22, 
       autopct = "%0.1f%%",
       pctdistance = 0.8,
       colors = ["#A7F107", "#61CDBB", "#CD61CD", "#CD617D", "#61CDBB"],
       labels = marcas_texto)

plt.title("Languages spoken in México", fontsize = 25) # Titulo del grafico.
plt.show()
```

#### Pie chart representig the percentage of the most important languages spoken in México
